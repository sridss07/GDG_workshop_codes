```python
import nltk
nltk.download("punkt")
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
nltk.download('stopwords')
stemmer = PorterStemmer()
text = "Chandragupta Maurya (350–295 BCE) was the founder of the Maurya Empire, which ruled over a geographically-extensive empire based in Magadha.He reigned from 320 BCE to 298 BCE."

sentences = nltk.sent_tokenize(text)
print(sentences)
print(sentences)
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [stemmer.stem(word) for word in words if word.lower() not in set(stopwords.words("english"))]
  sentences[i] = ' '.join(words)
print(sentences)
```

```python
import re
corpus = []

for i in range(len(sentences)):
  rev = re.sub('[^a-zA-Z]', ' ', sentences[i])
  rev = rev.lower()
  rev = rev.split()
  rev = [stemmer.stem(word) for word in rev if not word in set(stopwords.words('english'))]
  rev = ' '.join(rev)
  corpus.append(rev)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500)
X = cv.fit_transform(corpus).toarray()
```

```python
X
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer()
X = tfv.fit_transform(corpus).toarray()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

features = ['Sky', 'Wings', 'size']

bird_vector = np.array([3, 5, 4])
aeroplane_vector = np.array([5, 6, 8])

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0, 0, 0, bird_vector[0], bird_vector[1], bird_vector[2], color='r', label ='Bird', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, aeroplane_vector[0], aeroplane_vector[1], aeroplane_vector[2], color='b', label='Aeroplane', arrow_length_ratio=0.1)

ax.text(bird_vector[0], bird_vector[1], bird_vector[2], "Bird", color='r')
ax.text(aeroplane_vector[0], aeroplane_vector[1], aeroplane_vector[2], "Aeroplane", color='b')

ax.set_xlabel('Sky')
ax.set_ylabel('Wings')
ax.set_zlabel('Size')
ax.set_title('3D Representation')

ax.set_xlim(0, 6)
ax.set_ylim(0, 8)
ax.set_zlim(0, 10)

ax.legend()
plt.grid()
plt.show()
```
